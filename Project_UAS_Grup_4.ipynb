# Membaca Data

```python
import pandas as pd

url = "https://raw.githubusercontent.com/MairaNurul/Projects/refs/heads/main/Survei%20Pelanggan%20Coffee%20Shop%20(Jawaban).csv"

df = pd.read_csv(url, sep=';')
df
```

# PreProcessing


```python
# Daftar kolom yang mau disisakan (sudah diperbaiki agar sesuai dengan nama kolom di DataFrame)
kolom_dipakai = [
    "Program Studi",
    "Jenis Kelamin",
    "Dari mana anda mengetahui informasi suatu Coffee Shop?",
    "Coffee Shop 1",
    "Dari mana anda mengetahui informasi mengenai cafe/menu/diskon Coffee Shop pertama?",
    "Coffee Shop 2",
    "Dari mana anda mengetahui informasi mengenai cafe/menu/diskon Coffee Shop kedua?",
    "Coffee Shop 3",
    "Dari mana anda mengetahui informasi mengenai cafe/menu/diskon Coffee Shop ketiga?"
]

# Ambil hanya kolom tersebut
df_bersih = df[kolom_dipakai]
df_bersih
```

```python
# Daftar kolom jawaban untuk setiap coffee shop
media_cols = {
    "CoffeeShop1": "Dari mana anda mengetahui informasi mengenai cafe/menu/diskon Coffee Shop pertama?",
    "CoffeeShop2": "Dari mana anda mengetahui informasi mengenai cafe/menu/diskon Coffee Shop kedua?",
    "CoffeeShop3": "Dari mana anda mengetahui informasi mengenai cafe/menu/diskon Coffee Shop ketiga?"
}

# Proses jawaban per coffe shop  jadikan list
for cs, col in media_cols.items():
    df_bersih[col + "_list"] = df_bersih[col].apply(lambda x: [m.strip() for m in str(x).split(",")])

# Kumpulkan semua media unik
all_media = sorted({
    m
    for col in media_cols.values()
    for row in df_bersih[col + "_list"]
    for m in row
})

rows = []
for idx, row in df_bersih.iterrows():
    resp = f"R{idx+1}"

    for cs, col in media_cols.items():
        entry = {"Resp": resp}

        # isi kolom setiap media
        for m in all_media:
            entry[m] = m if m in row[col + "_list"] else ""

        # label coffee shop
        entry["Coffee Shop"] = cs
        entry["Utama"] = 3 if cs == "CoffeeShop1" else 2 if cs == "CoffeeShop2" else 1

        # ====== TAMBAHAN: isi kolom berdasarkan urutan Utama ======
        # Jika CoffeeShop1  ambil kolom asli Coffee Shop 1
        # Jika CoffeeShop2  ambil kolom asli Coffee Shop 2
        # Jika CoffeeShop3  ambil kolom asli Coffee Shop 3
        if entry["Utama"] == 3:
            entry["Coffee Shop Nama"] = df_bersih.loc[idx, "Coffee Shop 1"]
        elif entry["Utama"] == 2:
            entry["Coffee Shop Nama"] = df_bersih.loc[idx, "Coffee Shop 2"]
        else:
            entry["Coffee Shop Nama"] = df_bersih.loc[idx, "Coffee Shop 3"]
        # ==========================================================

        rows.append(entry)

final_df = pd.DataFrame(rows)
final_df
```

```python
# Mengambil semua kolom media (selain Resp, Coffee Shop, Utama, Coffee Shop Nama)
media_columns = [
    col for col in final_df.columns
    if col not in ["Resp", "Coffee Shop", "Utama", "Coffee Shop Nama"]
]

# Konversi: isi menjadi 1 jika ada teks, 0 jika kosong
for col in media_columns:
    final_df[col] = final_df[col].apply(
        lambda x: 1 if str(x).strip() != "" else 0
    )

final_df
```

### **1. Claimed : Hitung persentase "Dari mana anda mengetahui informasi suatu Coffee Shop?"**


```python
import pandas as pd
from collections import Counter

# Ambil kolom
col_awareness = "Dari mana anda mengetahui informasi suatu Coffee Shop?"
series = df_bersih[col_awareness].dropna()  # buang yang kosong

# Buat list semua media
all_media = []

for ans in series:
    # pecah jawaban jika ada lebih dari satu media (dipisah koma)
    medias = [m.strip() for m in str(ans).split(",") if m.strip() not in ["-", ""]]
    all_media.extend(medias)

# Hitung frekuensi tiap media
media_counts = Counter(all_media)

# Total responden
total_resp = len(df_bersih)

# Hitung persentase tiap media
media_percentages = {media: count / total_resp * 100 for media, count in media_counts.items()}

# Buat dataframe rapi
hasil_persentase = pd.DataFrame({
    "Jumlah": media_counts,
    "Persentase (%)": {k: round(v,2) for k,v in media_percentages.items()}
})

# Urutkan dari yang tertinggi
hasil_persentase = hasil_persentase.sort_values(by="Persentase (%)", ascending=False)
hasil_persentase
```

### **2. Derived Importance**

#### **Lakukan Regresi Logistik (Ordinal Logistic Regression)**

```python
!pip install mord
```

```python
import pandas as pd
import mord as m
from sklearn.metrics import accuracy_score, classification_report

# 1. Masukkan semua variabel X (media)
media_columns = [
    col for col in final_df.columns
    if col not in ["Resp", "Coffee Shop", "Utama", "Coffee Shop Nama"]
]

# Pastikan df = dataframe kamu
X = final_df[media_columns]

# 2. Kolom Y harus ordinal, misal 13
y = final_df["Utama"]   # target ordinal: 1,2,3

# 3. Model Ordinal Logistic Regression
model_ord = m.LogisticIT()
model_ord.fit(X, y)

# 4. Prediksi
y_pred = model_ord.predict(X)

import numpy as np

# 5. Output Koefisien dan Intercept
coefs = np.atleast_1d(model_ord.coef_)  # pastikan bisa di-iterasi

print("Koefisien per variabel:")
for var, coef in zip(media_columns, coefs):
    print(f"{var:35s} : {coef:.4f}")

print("\nIntercept (cutpoints _j):", model_ord.theta_, "\n")

# 6. Evaluasi Model
print("Akurasi:", accuracy_score(y, y_pred))
print("\nClassification Report:")
print(classification_report(y, y_pred))
```

#### **Hitung Odds Ratio tiap Faktor**

```python
import pandas as pd
import numpy as np

# Ambil koefisien dari model
coefs = np.atleast_1d(model_ord.coef_)

# Hitung Odds Ratio
odds_ratio = np.exp(coefs)

# Buat dataframe rapi
odds_df = pd.DataFrame({
    "Media": media_columns,
    "Coef": coefs,
    "Odds Ratio": odds_ratio
}).sort_values(by="Odds Ratio", ascending=False)

odds_df
```

#### **Lakukan Normalisasi**

```python
# Normalisasi Odds Ratio
odds_df["Normalized OR"] = odds_df["Odds Ratio"] / odds_df["Odds Ratio"].max()

# Urutkan kembali
odds_df = odds_df.sort_values(by="Normalized OR", ascending=False)
odds_df
```

### **3. Hitung nilai tengah masing-masing sebagai batas kuadran**

```python
# Hitung mid-point (batas kuadran)
# Mendefinisikan claimed_percentage dan X_vars untuk analisis kuadran
claimed_percentage = hasil_persentase["Persentase (%)"]
X_vars = media_columns

mid_performance = claimed_percentage.mean()
mid_importance = odds_df["Normalized OR"].mean()

mid_performance, mid_importance
```

```python
# 1. Filter & Align Claimed Percentage

# Sesuaikan index claimed_percentage dengan X_vars
claimed_percentage_filtered = claimed_percentage[claimed_percentage.index.isin(X_vars)]
claimed_percentage_aligned = claimed_percentage_filtered.reindex(X_vars)

# Hitung midpoint performance
mid_performance_aligned = claimed_percentage_aligned.mean()

# 2. Siapkan DataFrame Awal untuk Quadrant Analysis
quadrant_df = pd.DataFrame({
    "Media": media_columns,
    "Claimed_Percentage": claimed_percentage_aligned.values
})

# 2. Merge Importance (Odds Ratio) berdasarkan nama media
quadrant_df = quadrant_df.merge(
    odds_df[["Media", "Normalized OR"]],
    on="Media",
    how="left"
)

quadrant_df = quadrant_df.rename(columns={"Normalized OR": "Importance"})

# 3. Hitung nilai tengah masing-masing sebagai batas kuadran
mid_importance = 0.5  # Penyesuaian manual untuk visualisasi yang lebih baik, berdasarkan nilai mean sebelumnya: np.float64(0.5814961476042676)
mid_performance = 25  # Penyesuaian manual untuk visualisasi yang lebih baik, berdasarkan nilai mean sebelumnya: np.float64(27.596874999999997)

# 4. Tentukan Quadrant
def assign_quadrant(row):
    if row["Claimed_Percentage"] >= mid_performance and row["Importance"] >= mid_importance:
        return "Q1 - Maintain (Most Invest)"
    elif row["Claimed_Percentage"] < mid_performance and row["Importance"] >= mid_importance:
        return "Q2 - Improve (Opportunity to Optimize)"
    elif row["Claimed_Percentage"] < mid_performance and row["Importance"] < mid_importance:
        return "Q3 - Low Priority"
    else:
        return "Q4 - Occasionally (Good for new product)"

quadrant_df["Quadrant"] = quadrant_df.apply(assign_quadrant, axis=1)

# 5. Tampilkan Hasil
quadrant_df = quadrant_df.sort_values(by="Importance", ascending=False)
quadrant_df
```

### **4. Analisis Faktor dan Performa Faktor di Beberapa Brand**

Persebaran setiap kuadran berdasarkan matriks SoA :

*   Most Invest (Prioritas Investasi Utama) kuadran ini memiliki media dengan awareness dan derived interest yang tinggi, yaitu keluarga, Instagram, TikTok, dan Teman.
*   Opportunity to Optimize (Kesempatan untuk dioptimalkan) kuadran ini memiliki awareness yang rendah tetapi derived interest yang tinggi, yaitu ShopeeFood, blog, Iklan Online, X (Twitter), Event, Iklan Offline (Brosur/Pamflet).
*   Occasionally (Digunakan Sesekali) kuadran ini memiliki awareness yang tinggi tetapi derived interest yang rendah, yaitu Lihat langsung di lokasi dan GoFood.
*   Low Priority (Prioritas Rendah) kuadran ini memiliki awareness dan derived interest yang cukup rendah, yaitu Komunitas, GrabFood, Artikel, dan YouTube.
